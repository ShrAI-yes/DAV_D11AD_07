## **Python**

> Matplotlib is a comprehensive 2D plotting library that produces static, animated, and interactive visualizations in Python. It provides a wide variety of plots, including line plots, scatter plots, bar plots, histograms, and more and is highly customizable, allowing users to control aspects like colors, labels, and styles.

>  Seaborn is built on top of Matplotlib and provides a high-level interface for statistical data visualization. It simplifies the process of creating informative and attractive statistical graphics. It has specialized functions for statistical data visualization, such as heatmaps, violin plots, and pair plots along with built-in themes and color palettes for aesthetically pleasing visualizations. It also provides integration with Pandas data structures for seamless data handling.


```r
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pot
```

## Load Dataset

```r
data = pd.read_csv('/content/drive/MyDrive/Dataset/cars.csv')

data.dropna(axis=0,inplace = True)
```

```r
count = data['body_type'].value_counts()
labels = count.index
```

```r
pot.figure(figsize=(8,8))
pot.pie(count, labels= labels, autopct='%1.1f%%')
pot.title('Distribution of Car Body Type')
pot.axis('equal')
pot.show()
```

```r
counts = data['engine_fuel'].value_counts()

pot.figure(figsize=(8, 8))
ax = sns.countplot(x='engine_fuel', data=data, palette='viridis')

for i, count in enumerate(counts):
    ax.text(i, count + 0.1, str(count), ha='center', va='bottom')

pot.xlabel('Engine Fuels')
pot.ylabel('Count')
pot.title('Engine Fuel Types')

pot.grid(True)
pot.show()
```
